### Paso 1: Calcular la matriz TF-IDF para el corpus `data/01_corpus_turismo_500.txt`

```python
import nltk
nltk.download('stopwords')
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
import os

# Cargar el corpus
corpus_path = os.path.join('..', 'data', '01_corpus_turismo_500.txt')
with open(corpus_path, 'r', encoding='utf-8') as f:
    documents = [line.strip() for line in f if line.strip()]  # Cada línea es un documento

print(f"Número de documentos: {len(documents)}")
print("Ejemplo de documentos:")
for i in range(min(5, len(documents))):
    print(f"Documento {i+1}: {documents[i][:100]}...")

# Descargar stop words en español 
spanish_stopwords = stopwords.words('spanish')

# Crear el vectorizador TF-IDF
vectorizer = TfidfVectorizer(stop_words=spanish_stopwords, lowercase=True) 

# Ajustar y transformar los documentos
tfidf_matrix = vectorizer.fit_transform(documents)

# Obtener el vocabulario
vocab = vectorizer.get_feature_names_out()

# Convertir la matriz TF-IDF a un DataFrame para visualización
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=vocab)

print(f"Vocabulario único: {len(vocab)} términos")
print("Matriz TF-IDF (primeros 5 documentos y 10 términos):")
print(df_tfidf.iloc[:5, :10])  # Mostrar una muestra
```

### Paso 2: Construir el corpus `Gutenberg 1000`

El corpus `Gutenberg 1000` es un corpus compuesto por 1000 libros de Gutenberg Project

```python
import os

# Definir la ruta al corpus Gutenberg
gutenberg_dir = os.path.join('..', 'data', 'gutenberg', 'data')
file_names = sorted([f for f in os.listdir(gutenberg_dir) if f.lower().endswith('.txt')])

# Leer hasta 1000 archivos de texto
documents_gutenberg = []
doc_ids = []
for file_name in file_names[:1000]:
    file_path = os.path.join(gutenberg_dir, file_name)
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
        text = f.read().replace('\n', ' ').strip()
        if text:
            documents_gutenberg.append(text)
            doc_ids.append(file_name)

print(f"Archivos de Gutenberg encontrados: {len(file_names)}")
print(f"Documentos cargados en el corpus Gutenberg 1000: {len(documents_gutenberg)}")
print("Primeros 5 archivos del corpus:")
for name in doc_ids[:5]:
    print(f"- {name}")
```

### Paso 3: Calcular la matriz TF-IDF para el corpus `Gutenberg 1000`

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

vectorizer_gutenberg = TfidfVectorizer(
    stop_words='english',
    lowercase=True
    #max_df=0.95,
    #min_df=2,
    #norm='l2'
)

tfidf_gutenberg = vectorizer_gutenberg.fit_transform(documents_gutenberg)

vocab_gutenberg = vectorizer_gutenberg.get_feature_names_out()

print(f"Dimensiones de la matriz TF-IDF de Gutenberg: {tfidf_gutenberg.shape}")
print(f"Términos únicos en Gutenberg: {len(vocab_gutenberg)}")

sample_df = pd.DataFrame(tfidf_gutenberg[:5, :10].toarray(), columns=vocab_gutenberg[:10], index=doc_ids[:5])
print(sample_df)

first_doc = tfidf_gutenberg[0].toarray().ravel()
if first_doc.size > 0:
    top_idx = first_doc.argsort()[-10:][::-1]
    top_terms = [(vocab_gutenberg[i], first_doc[i]) for i in top_idx if first_doc[i] > 0]
    print("\nTop términos del primer documento:")
    for term, score in top_terms:
        print(f"{term}: {score:.4f}")
```

### Paso 4: Programar una función `buscar()` para el corpus `Gutenberg 1000`

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def buscar(query, top_n=10):

    # Procesar la consulta con el mismo vectorizador
    query_vec = vectorizer_gutenberg.transform([query])
    
    # Calcular similitud coseno entre la consulta y todos los documentos
    similarities = cosine_similarity(query_vec, tfidf_gutenberg).flatten()
    
    # Obtener los índices de los documentos más similares
    top_indices = np.argsort(similarities)[::-1][:top_n]
    
    # Crear lista de resultados: (doc_id, similitud)
    results = [(doc_ids[i], similarities[i]) for i in top_indices if similarities[i] > 0]
    
    return results

# Ejemplo de uso
query_example = "lord"
results = buscar(query_example, top_n=5)
print(f"Resultados para la consulta '{query_example}':")
for doc_id, sim in results:
    print(f"Documento: {doc_id}, Similitud: {sim:.4f}")
```
