## _Pasos:_
1. Leer el corpus

```python
import os

path = "..\\data\\gutenberg\\data"

archivos = os.listdir(path)[:100]  # solo los primeros 100

contenidos = {}
for archivo in archivos:
    with open(os.path.join(path, archivo), "r", encoding="utf-8") as f:
        contenidos[archivo] = f.read()
```

```python
import time

start_time = time.time()
query = "Mary"
resultados = []
for key in contenidos.keys():
    if query in contenidos[key]:
        resultados.append(key)

print(f"Los archivos que contienen la palabra '{query}' son: {resultados}")
end_time = time.time()
print(f"Tiempo de búsqueda: {end_time - start_time} segundos")
```

2. Definir una función de búsqueda_

```python
def buscar(query, contenidos):
    resultados = []
    for key in contenidos.keys():
        if query in contenidos[key]:
            resultados.append(key)
    return resultados
```

3. Separar todo el contenido en palabras

```python
palabras = set()
for contenido in contenidos.values():
    palabras.update(contenido.split())

print(len(palabras))
```

4. Usar la función y medir tiempos

```python
import time
# Buscar una palabra en vez de todas las palabras del vocabulario
start_time = time.time()
query = "Mary"
resultados = buscar(query, contenidos)
end_time = time.time()

print(f"Los archivos que contienen '{query}': {resultados}")
print(f"Tiempo de búsqueda: {end_time - start_time:.4f} segundos")
```
